```python
# 创建虚拟环境，由于python最大的问题是系统环境不同项目单独建立各自环境
conda create -p /root/autodl-tmp/project_env python=3.10 -y
```

```python
#激活创建的虚拟环境
!conda activate /root/autodl-tmp/project_env
```

```python
# 1. 安装内核工具
!pip install ipykernel
```

```python
#打开终端
conda activate /root/autodl-tmp/project_env
# 1. 安装内核工具
pip install ipykernel

# 2. 注册内核（名字可以叫 My_Project）
python -m ipykernel install --user --name project_env --display-name "Python (Project_Env)"

#重启内核，右上角切换为自建内核
#补充：
#如果终端里 conda activate 报错怎么办？
#在 AutoDL 的新机器上，经常会遇到 conda init 没配置好的问题，导致报错 CommandNotFoundError。
#解决办法（仅需在终端运行一次）：

# 1. 初始化 conda
conda init bash

# 2. 刷新终端环境（或者关掉终端重新开一个）
source ~/.bashrc

# 3. 现在再尝试激活
conda activate /root/autodl-tmp/project_env
```
